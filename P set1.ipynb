## Question 6
* For this problem you will build a small monte-carlo experiment. We will do this a lot in the class. The basic idea is to find the standard deviation for the mean for samples of size [10,25,100] of uniform random variables on the range zero to one, [0,1]. Use a big for loop to implement the iterations of the monte-carlo run. Do this 1000 times. For each iteration of the loop draw a set of uniform random variables corresponding to the above sample sizes. In each case find the mean, and store it in a vector. Once out of the loop, report the mean and standard deviations for the vectors generated in the for loop. Your program will be outputting 6 values at the end.

```python
# for sample size 10
s10=np.zeros([1000,10])
for i in range(0,1000):
    s10[i,:] = np.random.uniform(low=0,high=1,size=10)
m1=s10.mean(axis=1)
print(np.mean(m1))
print(np.std(m1))

# for sample size 25
s25=np.zeros([1000,25])
for i in range(0,1000):
    s25[i,:] = np.random.uniform(low=0,high=1,size=25)
m2=s25.mean(axis=1)
print(np.mean(m2))
print(np.std(m2))

# for sample size 100
s100=np.zeros([1000,100])
for i in range(0,1000):
    s100[i,:] = np.random.uniform(low=0,high=1,size=100)
m3=s100.mean(axis=1)
print(np.mean(m3))
print(np.std(m3))
```
